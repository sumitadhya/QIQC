```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector

# Create a quantum circuit
qc = QuantumCircuit(4)
qc.x(1)  # Apply Hadamard gate to qubit 0
qc.h(2)  # Apply X gate to qubit 1
qc.x(3)
qc.h(3)
# Get the statevector of the quantum circuit
state = Statevector.from_instruction(qc)

# Plot the state on multiple Bloch spheres
plot_bloch_multivector(state, title='Multi-Qubit States on Bloch Spheres')
```

For the state $|\psi\rangle=cos\frac{\theta}{2}|0\rangle+e^{i\phi}sin\frac{\theta}{2}|1\rangle$, the arrow in the Bloch sphere points by taking $\theta$ as the polar angle and $\phi$ as the azimuthal angle.


#### Qubit 0: 
Since no gates were applied to it, it remains in the ( $|0\rangle$ ) state, which is represented as a vector pointing to the north pole of the Bloch sphere.
#### Qubit 1: 
It is in the ( $|1\rangle$ ) state due to the X gate, which is represented as a vector pointing to the south pole of the Bloch sphere. 
#### Qubit 2: 
It is in a superposition state with phase zero ($\frac{|0\rangle + |1\rangle}{\sqrt{2}}$) due to the Hadamard gate, represented by a vector lying on the equator of the Bloch sphere. 
#### Qubit 3: 
It is in a superposition state ($\frac{|0\rangle - |1\rangle}{\sqrt{2}}$) with a relative phase of ( $\pi$ ) due to the X gate followed by a Hadamard gate. This would be represented by a vector lying on the equator of the Bloch sphere, pointing along the negative x-axis.

The first two are not in superposition so point along the poles.
The last two are in superposition so lie in equator.

```python
from qiskit import QuantumCircuit

# Create a quantum circuit with 3 qubits
qc = QuantumCircuit(3)

# Apply a Hadamard gate to the first qubit
qc.h(0)

# Apply CNOT gates to entangle the first qubit with the second and third qubits
qc.cx(0, 1)
qc.cx(0, 2)

# The circuit now prepares the state (|000⟩ + |111⟩) / √2
qc.draw('mpl')
```

```python
#for any n.

n=9

from qiskit import QuantumCircuit

# Create a quantum circuit with n qubits
qc = QuantumCircuit(n)

# Apply a Hadamard gate to the first qubit
qc.h(0)

# Apply CNOT gates to entangle the first qubit with the remaining n-1 qubits
for qubit in range(1,n):
    qc.cx(0, qubit)


# The circuit now prepares the state (|0...0⟩ + |1...1⟩) / √2
qc.draw('mpl')
```
